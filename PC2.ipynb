```python
import pandas as pd

tablelink='https://raw.githubusercontent.com/Controles-ICDERebatta/PC_2/refs/heads/main/HDI.xlsx%20-%20Table%201.csv'

dataPC2=pd.read_csv(tablelink)
dataPC2
```

```python
print(dataPC2.columns)
```

```python
keep=['Back','Table 1. Human Development Index and its components ', 'Unnamed: 2', 'Unnamed: 4', 'Unnamed: 6', 'Unnamed: 8', 'Unnamed: 10', 'Unnamed: 12', 'Unnamed: 14']

data_firstcleaned=dataPC2.loc[:,keep]
data_firstcleaned
```

```python
change={'Back': "Order rank", 'Unnamed: 2':"Human_Development_Index_HDI", 'Unnamed: 4':"Life expectancy at birth SDG3",'Unnamed: 6':"Expected years of schooling SDG4_3", 'Unnamed: 8':"Many years of schooling SDG4_4", 'Unnamed: 10':"GNI per capita SDG8_5", 'Unnamed: 12':"GNI per capita rank minus HDI rank", 'Unnamed: 14':"HDI rank"}
data_firstcleaned.rename(columns=change,inplace=True)
data_firstcleaned

```

```python
objectCols=data_firstcleaned.select_dtypes(include=['object']).columns
objectCols
```

```python
data_firstcleaned.columns=data_firstcleaned.columns.str.replace(r'\s','_',regex=True)
data_firstcleaned
```

```python
data_firstcleaned.dropna()
```

```python
data_secondcleaned=data_firstcleaned.dropna()
data_secondcleaned
```

```python
#data_secondcleaned['Human_Development_Index_(HDI)']=data_secondcleaned['Human_Development_Index_(HDI)'].replace(',','.', regex=True)
#data_secondcleaned

data_secondcleaned.loc[:, 'Human_Development_Index_HDI'] = data_secondcleaned.loc[:, 'Human_Development_Index_HDI'].str.replace(',', '.', regex=False)
data_secondcleaned.loc[:, 'Life_expectancy_at_birth_SDG3'] = data_secondcleaned.loc[:, 'Life_expectancy_at_birth_SDG3'].str.replace(',', '.', regex=False)
data_secondcleaned.loc[:, 'Expected_years_of_schooling_SDG4_3'] = data_secondcleaned.loc[:, 'Expected_years_of_schooling_SDG4_3'].str.replace(',', '.', regex=False)
data_secondcleaned.loc[:, 'Many_years_of_schooling_SDG4_4'] = data_secondcleaned.loc[:, 'Many_years_of_schooling_SDG4_4'].str.replace(',', '.', regex=False)
data_secondcleaned.loc[:, 'GNI_per_capita_SDG8_5'] = data_secondcleaned.loc[:, 'GNI_per_capita_SDG8_5'].str.replace(',', '.', regex=False)
data_secondcleaned.loc[:, 'GNI_per_capita_rank_minus_HDI_rank'] = data_secondcleaned.loc[:, 'GNI_per_capita_rank_minus_HDI_rank'].str.replace(',', '.', regex=False)
#data_secondcleaned.loc[:, 'HDI_rank'] = data_secondcleaned.loc[:, 'HDI_rank'].str.replace(',', '.', regex=False)

data_secondcleaned
```
